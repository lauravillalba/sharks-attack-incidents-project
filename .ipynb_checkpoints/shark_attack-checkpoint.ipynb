```python
#Se eliminan las siguientes columnas: pdf, href values, href, original order, Unnamed: 22 y Unnamed 23.

df_minColumns= df.drop (['pdf', 'href formula', 'href','original order', 'Unnamed: 22','Unnamed: 23'],axis=1)
df_minColumns.head(1)
```

```python
#Se comprueba si los valores de Case Number son únicos:

df_minColumns['Case Number'].is_unique
```

```python
#Número de duplicados:

len(df_minColumns[df_minColumns['Case Number'].duplicated(keep=False)])
```

```python
#Consideración: Case Number debe ser un campo de valores únicos. Se eliminan todos los Case Number duplicados.

df_nonDup = df_minColumns.drop_duplicates('Case Number', keep=False)
df_nonDup.head(1)
```

```python
#Depurando campo Year:
df_nonDup.groupby("Year").agg({"Year":"count"})

df_nonDup["500años"]=pd.cut(df_nonDup.Year, bins=[-500,0,500,1000,1500,2000,2500], labels= range(-500,2500,500))
```

```python
df_nonDup[["500años","Year"]]["500años"].value_counts()
```

```python
#Se sacan de la muestra los accidentes anteriores al año 1900 y se renombra la columna "500años" --> "decada"
df_years=df_nonDup[(df_nonDup.Year>=1900)]

df_years.rename(columns={'Case Number': 'caseNumber','Sex ':'Sex','Species ':'Species', "500años": "Decada"},inplace=True)

df_years["Decada"]=pd.cut(df_years.Year, bins=[1899,1909,1919,1929,1939,1949,1959,1969,1979,1989,1999,2009,2019], 
                          labels= range(1900,2020,10))
```

```python
df_years.groupby("Decada").agg({"Year":"count"})
```

```python
df_years.columns
```

```python
# Reorganización de columnas:

df_years = df_years[["caseNumber", "Decada","Year","Date","Time", "Country","Area","Location","Name", "Sex","Age", 
                     "Activity","Injury", "Fatal (Y/N)","Species"]]

df_years = df_years.sort_values(by=["Year"], ascending=False).reset_index(drop=True)
df_years.head(3)
```

```python
#Extracción de la fecha del 'caseNumber':

dates_info={
            'caseNumber': list(df_years["caseNumber"])
            }
caseNumber= list(df_years["caseNumber"])
dates_info['newDate'] = []

for dat in caseNumber:
    
    res = re.findall(r'\d{4}\.\d{2}\.\d{2}', dat)
    
    if res:
        dates_info["newDate"].append(res[0])
    else:
        dates_info["newDate"].append('')
```

```python
#Extracción del año del 'caseNumber':

dates_info['year_newDate']=[]

for year in caseNumber:
    res=re.findall(r'^\d{4}',year)
    if res:
        dates_info['year_newDate'].append(res[0])
    else:
        dates_info['year_newDate'].append('')
```

```python
#Extracción del mes de 'newDate'-->se utilizará para establecer la estación:

dates_info['month']=[]

for mon in dates_info["newDate"]:
    res= re.findall(r'\.(\d{2})\.', mon)
    
    if res:
        dates_info['month'].append(res[0])
            
    else:
        dates_info['month'].append('')
        
```

```python
# Llevamos newDate y month al dataframe df_year:

df_newDate = pd.DataFrame(dates_info)
df_newDate = pd.merge(df_years,df_newDate, on='caseNumber')
```

```python
#Comprobación Year (origen) == year_newDate (extracción de caseNumber) para mantener una única columna:
df_newDate.columns

def fn (rowSeries):
    if int(rowSeries.Year)==int(rowSeries.year_newDate):
        return True
    return False
    
df_newDate[['Year', 'year_newDate']].apply(fn,axis=1, result_type='broadcast').groupby('Year').sum()
```

```python
df_newDate.columns
```

```python
#Reestructuración de columnas:

df_newDate= df_newDate[["caseNumber","newDate", "Decada","Year", "month", "Time", "Country","Area","Location","Name", "Sex","Age", 
                     "Activity","Injury", "Fatal (Y/N)","Species"]]

df_newDate=df_newDate.rename(columns={"Decada":"decada", "Year":"year", "Time":"time", "Country": "country","Area": "area", "Location":"location", "Name":"name", "Sex":"sex", "Age": "age",
                      "Activity":"activity","Injury":"injury", "Fatal (Y/N)":"fatal","Species":"species"})
df_newDate.head(3)
```

```python
#Top 10 --> Países con más ataques:
attacksPerCountry = df_newDate.groupby('country').agg ({'caseNumber':'count'}).sort_values(by=['caseNumber'], ascending=False)
attacksPerCountry.head(10)
```

```python
#Hipótesis: Hay más ataques en verano que en invierno, pero en invierno los ataques son mayormente fatales 
# =======> "No lo cuentas" 
#Se analizará esta hipótesis sobre el Top10 de países.
```

```python
df_countries = df_newDate[(df_newDate.country.isin(['USA','AUSTRALIA','SOUTH AFRICA','PAPUA NEW GUINEA','BRAZIL', 'NEW ZELAND',
                                                 'BAHAMAS', 'MEXICO', 'ITALY','REUNION']))]
```

```python
#Revisión atributo 'fatal':

df_countries.isnull().sum()
df_countries[(~df_countries.fatal.isin(['Y','N']))]

df_countries.groupby('fatal').agg({'caseNumber':'count'})

def transforma_fatal (n):
    if n== 'Y' or n=='N':
        return n
    else:
        return 'NaN'

df_countries['fatal']=df_countries['fatal'].apply(transforma_fatal)
```

```python
#Nos quedamos con las columnas necesarias:
df_cleaned = df_countries[['caseNumber', 'newDate', 'decada', 'year', 'month', 'time', 'country',
                            'sex', 'age', 'activity', 'fatal']]
```

```python
#Añadimos un atribudo 'hemisphere' que asignamos por país:

def hemisphere(x):
    if x in ['USA','BAHAMAS', 'MEXICO', 'ITALY']:
        return 'NORTE'
    else:
        return 'SUR'

df_cleaned['hemisphere']=df_cleaned['country'].apply(hemisphere)

df_cleaned.head(3)
```

```python
#Añadimos nuevo atributo para diferenciar meses de frío y calor según mes y hemisferio:

def season(month,hemisphere):
    if month in ['04','05','06','07','08','09'] and hemisphere == 'NORTE':
        return 'WARM'
    elif month in ['01','02','03','10','11','12'] and hemisphere == 'SUR':
        return 'WARM'
    else:
        return 'COLD'    

df_cleaned['season']=df_cleaned.apply(lambda x: season(x['month'], x['hemisphere']), axis=1)

df_cleaned
```

```python
df_resum=df_cleaned[['caseNumber', 'decada', 'hemisphere', 'season', 'fatal']]
df_resum
```

```python
#Comprobación de la primera hipótesis --> Hay más ataques de tiburón durante los meses cálidos que durante los fríos:

df_resum.groupby('season').agg ({'caseNumber':'count'
                                }).sort_values(by=['caseNumber'], ascending=False)
```

```python
q = df_resum.groupby("season").agg({
                                    "caseNumber":"count"
})
q.plot.bar()  
```
