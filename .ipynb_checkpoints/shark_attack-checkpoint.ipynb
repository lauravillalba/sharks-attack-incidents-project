-----------------------------
Analizando el resultado anterior, vamos a sacar de la muestra los ataques registrados antes del año 1900. 

Actualizaremos el nombre de varias columnas:'Case Number' por 'caseNumber', 'Sex ' por 'Sex', 'Species ' por 'Species' y '500años' por 'Decada'.


```python
df_years=df_nonDup[(df_nonDup.Year>=1900)]

df_years.rename(columns={'Case Number': 'caseNumber','Sex ':'Sex','Species ':'Species', 
                         "500años": "Decada"},inplace=True)

df_years["Decada"]=pd.cut(df_years.Year, bins=[1899,1909,1919,1929,1939,1949,1959,1969,1979,1989,1999,2009,2019], 
                          labels= range(1900,2020,10))

df_years.groupby("Decada").agg({"Year":"count"})
```

```python
df_years.columns
```

----------------------------------
Reorganizamos las columnas, las ordenamos por 'Year' y reseteamos el índice.

```python
df_years = df_years[["caseNumber", "Decada","Year","Date","Time", "Country","Area","Location","Name", "Sex","Age", 
                     "Activity","Injury", "Fatal (Y/N)","Species"]]

df_years = df_years.sort_values(by=["Year"], ascending=False).reset_index(drop=True)
df_years.head()
```

-----------------------------------
'caseNumber' viene dado por la fecha del ataque. Utilizaremos este dato para estraer la fecha, el año y el mes.

Creamos un diccionario que sólo contenga 'caseNumber' y le vamos a ir añadiendo 'newDate', 'year_newDate' y 'month'.


```python
#Extracción de la fecha del 'caseNumber':

dates_info={
            'caseNumber': list(df_years["caseNumber"])
            }
caseNumber= list(df_years["caseNumber"])
dates_info['newDate'] = []

for dat in caseNumber:
    
    res = re.findall(r'\d{4}\.\d{2}\.\d{2}', dat)
    
    if res:
        dates_info["newDate"].append(res[0])
    else:
        dates_info["newDate"].append('')

```

```python
#Extracción del año del 'caseNumber':

dates_info['year_newDate']=[]

for year in caseNumber:
    res=re.findall(r'^\d{4}',year)
    if res:
        dates_info['year_newDate'].append(res[0])
    else:
        dates_info['year_newDate'].append('')
```

```python
#Extracción del mes de 'newDate':

dates_info['month']=[]

for mon in dates_info["newDate"]:
    res= re.findall(r'\.(\d{2})\.', mon)
    
    if res:
        dates_info['month'].append(res[0])
            
    else:
        dates_info['month'].append('')
        
```

------------------------
Convertimos ahora el diccionario dates_info en DataFrame y luego lo unimos al nuestro ('df_newyear').

```python
df_newDate = pd.DataFrame(dates_info)
df_newDate = pd.merge(df_years,df_newDate, on='caseNumber')

df_newDate.head(3)
```

----------------------------------
Hacemos una comprobación entre las dos columnas de años ('Year' y 'year_newDate'). Si resultan iguales, nos quedaremos sólo con una. 
Para esto hemos creado la función 'checkYear'.

```python
df_newDate[['Year', 'year_newDate']].apply(fn.checkYear,axis=1, result_type='broadcast').groupby('Year').sum()
```

```python
df_newDate.columns
```

------------------------------------
Nuevamente reestructuramos las columnas y actualizamos todos los nombres para dejarlos en minúsculas:

```python
df_newDate= df_newDate[["caseNumber","newDate", "Decada","Year", "month", "Time", "Country","Area","Location","Name", "Sex","Age", 
                     "Activity","Injury", "Fatal (Y/N)","Species"]]

df_newDate=df_newDate.rename(columns={"Decada":"decada", "Year":"year", "Time":"time", "Country": "country","Area": "area", "Location":"location", "Name":"name", "Sex":"sex", "Age": "age",
                      "Activity":"activity","Injury":"injury", "Fatal (Y/N)":"fatal","Species":"species"})
df_newDate.head()
```

-------------------------------
Creamos un nuevo df con el top10

```python
df_countries = df_newDate[(df_newDate.country.isin(['USA','AUSTRALIA','SOUTH AFRICA','PAPUA NEW GUINEA','BRAZIL',
                                                    'NEW ZELAND','BAHAMAS', 'MEXICO', 'ITALY','REUNION']))]
df_countries.head(3)
```

-------------------------------
Revisamos la columna 'fatal' y podemos ver que se obtienen 4 valores null. Sin embargo al visualizar la tabla de valores distintos de 'Y' y 'N' se puede apreciar que existen otros no reconocidos como null pero que siguen siendo desconocidos. 

```python
display(df_countries.isnull().sum())
display(df_countries[(~df_countries.fatal.isin(['Y','N']))])
```

Limpiamos ahora la columna 'fatal' con la función 'transforma_fatal'. Con esta función se mantienen los valores 'Y' y 'N', y el resto de valores los reemplazamos por 'NaN'.

```python
df_countries['fatal']=df_countries['fatal'].apply(fn.transforma_fatal)

df_countries.groupby('fatal').agg({'caseNumber':'count'})
```

Generamos ahora un nuevo df reduciendo el número de columnas.

```python
df_cleaned = df_countries[['caseNumber', 'newDate', 'decada', 'year', 'month', 'time', 'country',
                            'sex', 'age', 'activity', 'fatal']]

display(df_cleaned.head(3))
```

-----------------------------------------------
Para poder analizar los datos atendiendo a la época del año necesitamos asignar a cada caso en qué hemisferio tiene lugar. Para esto hemos creado la función 'hemisphere', que según el país le asigna 'NORTE' o 'SUR'.

```python
df_cleaned['hemisphere']=df_cleaned['country'].apply(fn.hemisphere)

df_cleaned.head(3)
```

----------------------------
Ahora vamos a añadir a cada caso, la temporada en la que sucede. Para esto hemos creado la función 'season' que tiene en cuenta el mes y el hemisferio de cada caso.

```python
#Añadimos nuevo atributo 'season' para diferenciar meses fríos y cálidos según mes y hemisferio:

df_cleaned['season']=df_cleaned.apply(lambda x: fn.season(x['month'], x['hemisphere']), axis=1)

display(df_cleaned.head(3))
```

------------------
Creamos un nuevo df resumen.

```python
df_resum=df_cleaned[['caseNumber', 'decada', 'hemisphere', 'season', 'fatal']]
display(df_resum.head(3))
```

----------------------------------
Comprobamos la primera parte de la hipótesis:
        "HAY MÁS ATAQUES DE TIBURÓN DURANTE LOS MESES CÁLIDOS QUE DURANTE LOS FRÍOS"

```python
df_resum.groupby('season').agg ({'caseNumber':'count'
                                }).sort_values(by=['caseNumber'], ascending=False)
```

```python
q = df_resum.groupby("season").agg({
                                    "caseNumber":"count"
})

graf_season=q.plot.bar(color='darkblue') 
graf_season_title=plt.title('ATTACKS PER SEASON')
```

------------------------------------
Comprobamos la segunda parte de la hipótesis: "SI RECIBES UN ATAQUE DE TIBURÓN DURANTE LOS MESES FRÍOS ES MÁS PROBABLE QUE SEA FATAL"

```python
df_cold= df_resum[(df_resum.season == 'COLD')].groupby('fatal').agg ({'caseNumber':'count'})

#print('\n', '% fatal attacks in cold season: ',df_cold['caseNumber']['Y']/df_cold['caseNumber'].sum())

type_attack=['fatal','no_fatal','NaN']
num_attack=[df_cold['caseNumber']['Y'],df_cold['caseNumber']['N'], df_cold['caseNumber']['NaN']]

explode=[0.3,0,0]

colores=['royalblue','lightsteelblue','lightgrey']
                  
plt.pie(num_attack,labels=type_attack,explode=explode, autopct="%0.1f %%", colors=colores)

plt.title('FATAL ATTACKS IN COLD SEASON')
plt.show()
```

```python
df_warm=df_resum[(df_resum.season=='WARM')].groupby('fatal').agg({'caseNumber': 'count'})

#print('\n', '% fatal attacks in warm season: ',df_warm['caseNumber']['Y']/df_warm['caseNumber'].sum())

type_attack=['fatal','no_fatal','NaN']
num_attack=[df_warm['caseNumber']['Y'],df_warm['caseNumber']['N'], df_warm['caseNumber']['NaN']]

explode=[0.3,0,0]

colores=['darkred', 'rosybrown','mistyrose']
                  
plt.pie(num_attack,labels=type_attack,explode=explode, autopct="%0.1f %%", colors=colores)

plt.title('FATAL ATTACKS IN WARM SEASON')
plt.show()
```
